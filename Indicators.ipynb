```python
data = pd.read_csv('YNDX.csv')
data
```

```python
for i in (3, 5, 7, 10, 15, 20, 25, 30, 45, 60):
    win=i
    SMA = ta.trend.sma_indicator(data['OPEN'], window=win)
    EMA = ta.trend.ema_indicator(data['OPEN'], window=win)
    ATR = ta.volatility.average_true_range(data['HIGH'], data['LOW'], data['CLOSE'], window=win)
    ADX = ta.trend.adx(data['HIGH'], data['LOW'], data['CLOSE'], window=win)
    CCI = ta.trend.cci(data['HIGH'], data['LOW'], data['CLOSE'], window=win)
    RSI = ta.momentum.rsi(data['CLOSE'], window=win) 
    Williams = ta.momentum.williams_r(data['HIGH'], data['LOW'], data['CLOSE'], lbp=win)
    result = pd.concat([data, SMA, EMA,ATR,ADX,CCI,RSI,Williams], axis=1)
    win = str(win)
    result = result[result.index >= 60]
    result.to_csv('SNGS' + win + '.csv', index=False)
```
