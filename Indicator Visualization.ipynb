```python
import matplotlib.pyplot as plt
# Setting the figure size 
plt.rcParams["figure.figsize"] = (30,30)

# Subdividing the figure into 3 figures stacked in 1 row
fig, ax = plt.subplots(4, 3)

# First Plot - S&P 500 prices against time
ax[0,0].plot(data['CLOSE'][-100:], color = "blue", label = "IMOEX")
ax[0,0].set_title("Доходность", size = 24)
ax[0,0].legend()

# Second Plot - S&P 500 returns against time
ax[0,1].plot(data['sma_7'][-100:], color = "blue")
ax[0,1].set_title("SMA", size = 24)
ax[0,1].legend()

# Third Plot - S&P 500 volatility against time
ax[0,2].plot(data['ema_7'][-100:], color = "blue")
ax[0,2].set_title("EMA", size = 24)
ax[0,2].legend()

# First Plot - S&P 500 prices against time
ax[1,0].plot(data['atr'][-100:], color = "blue")
ax[1,0].set_title("ATR", size = 24)
ax[1,0].legend()

# Second Plot - S&P 500 returns against time
ax[1,1].plot(data['adx'][-100:], color = "blue")
ax[1,1].set_title("ADX", size = 24)
ax[1,1].legend()

# Third Plot - S&P 500 volatility against time
ax[1,2].plot(data['cci'][-100:], color = "blue")
ax[1,2].set_title("CCI", size = 24)
ax[1,2].legend()

# First Plot - S&P 500 prices against time
ax[2,0].plot(data['rsi'][-100:], color = "blue")
ax[2,0].set_title("RSI", size = 24)
ax[2,0].legend()

# Second Plot - S&P 500 returns against time
ax[2,1].plot(data['wr'][-100:], color = "blue")
ax[2,1].set_title("W%R", size = 24)
ax[2,1].legend()

# Third Plot - S&P 500 volatility against time
ax[2,2].plot(data['roc'][-100:], color = "blue")
ax[2,2].set_title("ROC", size = 24)
ax[2,2].legend()

# First Plot - S&P 500 prices against time
ax[3,0].plot(data['stochrsi_d'][-100:], color = "blue")
ax[3,0].set_title("%D", size = 24)
ax[3,0].legend()

# Second Plot - S&P 500 returns against time
ax[3,1].plot(data['stochrsi_k'][-100:], color = "blue")
ax[3,1].set_title("%K", size = 24)
ax[3,1].legend()

# Third Plot - S&P 500 volatility against time
ax[3,2].plot(data['fi_7'][-100:], color = "blue")
ax[3,2].set_title("FI", size = 24)
ax[3,2].legend()

# Used to display the plot free from any additional text in Jupyter notebooks
plt.show()
```
