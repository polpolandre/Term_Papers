```python
from datetime import datetime
def timesplit(datetime_):
    date, time = datetime_.split()
    year = int(date[:4])
    month = int(date[4:6])
    day = int(date[6:])
    hour = int(time[:2])
    minute = int(time[2:4])
    sec = int(time[4:])
    date_ = datetime(year = year,
                   month = month,
                   day = day,
                   hour = hour,
                   minute = minute,
                   second = sec)
    return date_
```

```python
data['DATETIME_new'] = data['DATETIME'].apply(timesplit)
```
